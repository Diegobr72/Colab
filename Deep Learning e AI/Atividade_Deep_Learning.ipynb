## 1) Crie uma função que receba dois números, x, y | x <y. A função deve imprimir, sequencialmente, o valor de x até y, calculando a soma de todos os números anteriores. Exemplo: dado x = 2 e y = 5, o programa deve imprimir:

<p>

### x = 2, soma = 2

<p>

### x = 3, soma = 5

<p>

### x = 4, soma = 9

<p>

### x = 5, soma = 14

<p>

```python
def f(x,y):
  soma = 0
  for i in range(y-x+1):
    soma += x
    print(f"x = {x}, soma = {soma}")
    x += 1
```

```python
f(2,5)
```

## **Desafio Extra devido o (Diego cabeção)**

```python
def fibonacci(x):
    a = 0
    b = 1
    for i in range(x):
        print(a)
        b = a + b
        a = b - a
```

```python
fibonacci(10)
```

## 2) Dada uma lista de inteiros, verifique se a soma de todos os elementos da lista, exceto o último, é maior que o valor do último elemento

```python
def soma_inteiros(x):
    soma = sum(x) - x[-1]

    if soma > x[-1]:
        print("A soma é maior que o último número")
    else:
        print("A soma não é maior que o último número")
```

```python
a = [4,5,2,6]
soma_inteiros(a)
```

## 4) Crie uma função que, dado um número X, determine se o mesmo é um palíndromo.

<p>

### Exemplo: 1221 é palíndromo

<p>

### 1222 não é palíndromo

```python
def palindromo(x):

  numero = 0
  inverso = 0
  numero = x

  while numero != 0:
    digito = numero % 10
    inverso = inverso * 10 + digito
    numero = numero // 10

  if x == inverso:
    print(f"O numero {x} é um palíndromo")

  else:
    print(f"O numero {x} não é palíndromo")
```

```python
palindromo(5115)

palindromo(34567)
```

## 5) Crie uma função que receba duas listas de inteiros positivos. A função deve criar e retornar uma terceira lista composta dos números pares da primeira lista e dos ímpares da segunda lista.

```python
def listasConbinadas(x,y):
  a = []

  for i in x:
    if i % 2 == 0:
      a.append(i)

  for i in y:
    if i % 2 != 0:
      a.append(i)

  print(a)

```

```python
lista1 = [98,765,4,32,123,45]
lista2 = [2,3,4,5,67,76,5,43]

listasConbinadas(lista1,lista2)
```

## 7) Dada uma lista de inteiros positivos, determine o segundo maior valor existente na lista.

```python
#Esse foi o codigo que foi aceito em aula

def segundo_Maior_numero(x):
  x.sort()
  a = x[-2]
  print(a)
```

```python
#Esse foi o codigo que você queria

def segundo_Maior_unico(x):
  x = list(set(x))
  x.sort()
  print(x[-2])
```

```python
listaA = [8,5,-1,6,1,23,53,2,62,62,10,10]

print("segundo maior numero da lista:")
segundo_Maior_numero(listaA)
print("segundo maior numero da lista sendo unico:")
segundo_Maior_unico(listaA)
```
